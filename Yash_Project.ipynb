```python
df.dropna(inplace = True)
```

```python
#boxplot
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['TotalPrice'])
plt.title('Boxplot of Total Price')
plt.xlabel('Total Price')
plt.show()

```

```python
#bar graph
plt.figure(figsize=(6, 4))
df['Product'].value_counts().plot(kind='bar', color='lightgreen')
plt.title('Number of Loans per Product')
plt.xlabel('Product')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
#scatter plot
plt.figure(figsize=(6, 4))
sns.stripplot(x='Region', y='TotalPrice', data=df, jitter=True)
plt.title('Total Price Distribution by Region')
plt.xlabel('Region')
plt.ylabel('Total Price')
plt.xticks(rotation=45)
plt.show()
```

```python
#pie chart
plt.figure(figsize=(6, 6))
df['CustomerGender'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Customer Gender Distribution')
plt.ylabel('')
plt.show()
```
